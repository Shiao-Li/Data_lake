```python
#Librerias
import csv
import couchdb
import pandas as pd
```

```python
# Leer archivo csv
dfcouchdb1 = pd.read_csv('cause_of_deaths.csv', encoding='utf-8')
```

```python
#Impresion del dataframe
dfcouchdb1
```

```python
#Contabilizar datos nulos
dfcouchdb1.isnull().sum()
```

```python
#Convertir el dataframe en un diccionario
datoscouchdb1 = dfcouchdb1.to_dict(orient='records')

#Conexion couchdb
servidorcouchdb1 = couchdb.Server('http://admin:admin@localhost:5984/')
dbcouchdb1 = servidorcouchdb1['muertes_mundiales1']

#Ingresar los datos en la base de datos
for documento in datoscouchdb1:
    dbcouchdb1.save(documento)
```
